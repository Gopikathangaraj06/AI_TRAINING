```python
import pandas as pd #data analysis
import numpy as np #data manipulation mathamatical calc
import matplotlib.pyplot as plt #data visualisation
import seaborn as sns #advanced data visualise higher level
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df=pd.read_csv("/content/seattle-weather.csv")
df.head()
```

```python
df.isnull()
```

```python
featured=['precipitation','temp_min','wind']
x=df[featured]
y=df['temp_max']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred =model.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
plt.figure(figsize=(15,5)) # Adjusted figsize for 3 plots

plt.subplot(1,3,1)
sns.scatterplot(x=x['temp_min'], y=y, color='orange', label='WEATHER PREDICTION')
plt.title("temp_min vs temp_max")
plt.xlabel("TEMP_MIN")
plt.ylabel("TEMP_MAX")
plt.legend()
plt.grid(True)

plt.subplot(1,3,2)
sns.scatterplot(x=x['precipitation'], y=y, color='blue', label='WEATHER PREDICTION')
plt.title("precipitation vs temp_max")
plt.xlabel("PRECIPITATION")
plt.ylabel("TEMP_MAX")
plt.legend()
plt.grid(True)

plt.subplot(1,3,3)
sns.scatterplot(x=x['wind'], y=y, color='green', label='WEATHER PREDICTION')
plt.title("wind vs temp_max")
plt.xlabel("WIND")
plt.ylabel("TEMP_MAX")
plt.legend()
plt.grid(True)

plt.tight_layout() # Adjusts subplot params for a tight layout
plt.show()
```

```python
from sklearn.linear_model import SGDRegressor
model = SGDRegressor(max_iter=1000, learning_rate="constant", eta0=0.0001, warm_start=True)

losses = []
epochs = 100

for epoch in range(epochs):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    loss = mean_squared_error(y_test, y_pred)
    losses.append(loss)
    print(f"Epoch {epoch+1}/{epochs} - Loss: {loss}")
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
print(mean_squared_error(y_test, lr.predict(x_test)))
```
